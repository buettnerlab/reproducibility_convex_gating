```python
def make_pivot_table_hypergate(df_hypergate,dataset = 'large PBMC', cell_type_key = 'cluster cell_type_lvl_1'):
    df_hypergate = df_hypergate.rename(columns={'cell_type': 'celltype'})
    df_hypergate['dataset'] = dataset
    df_hypergate['model'] = 'Hypergate'
    df_hypergate = df_hypergate.pivot_table(
    index=['sample', 'celltype', 'set_size', 'ratio', 'train_ratio', 'time', 'model','dataset'],
    columns='score',
    values='value'
        ).reset_index()
    df_hypergate = df_hypergate.drop(columns=['train_ratio'])
    df_hypergate.columns.name = None
    df_hypergate['cell_type_key'] = cell_type_key
    df_hypergate = df_hypergate.rename(columns={'set_size': 'set'})
    return df_hypergate
```

```python
def plot_scoretype_stackedbar(ax, df, score_type, title):
    # Determine winner
    df['winner'] = np.where(
        df['ConvexGating'] > df['Hypergate'], 'ConvexGating higher',
        np.where(df['ConvexGating'] < df['Hypergate'], 'Hypergate higher', 'Equal')
    )

    # Count wins per dataset_celltypekey
    count_df = df.groupby(['dataset_celltypekey', 'winner']).size().unstack(fill_value=0)

    # Ensure both model columns exist
    for col in ['ConvexGating higher', 'Hypergate higher']:
        if col not in count_df.columns:
            count_df[col] = 0

    # Compute percentages
    percentage_df = count_df[['ConvexGating higher', 'Hypergate higher']].div(
        count_df.sum(axis=1), axis=0
    ) * 100

    # Plot stacked bar
    percentage_df.plot(
        kind='bar',
        stacked=True,
        ax=ax,
        color=[custom_palette['ConvexGating'], custom_palette['Hypergate']]
    )

    # Add horizontal reference line at 50%
    ax.axhline(50, color='k', linestyle='-', linewidth=1,alpha=0.8)

    # Axis labels and title
    ax.set_ylabel("Percentage")
    ax.set_title(title)
    ax.set_ylim(0, 105)

    # Move legend outside
    ax.legend(
        bbox_to_anchor=(1.02, 1),
        loc='upper left',
        borderaxespad=0
    )

# Example usage
fig, axes = plt.subplots(3, 1, figsize=(8, 8), sharex=True)

plot_scoretype_stackedbar(axes[0], df_pivot_1to1[df_pivot_1to1['score_type'] == 'precision'], 'precision', "Precision")
plot_scoretype_stackedbar(axes[1], df_pivot_1to1[df_pivot_1to1['score_type'] == 'recall'], 'recall', "Recall")
plot_scoretype_stackedbar(axes[2], df_pivot_1to1[df_pivot_1to1['score_type'] == 'f1'], 'f1', "F1")
plt.savefig(os.path.join(base_save_path_rev12, pre + '_stackedbar_Hypergate_CG_winner_1to1.pdf'), bbox_inches='tight')
plt.show()
```
