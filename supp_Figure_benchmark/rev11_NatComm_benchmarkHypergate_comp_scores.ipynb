```python
def make_pivot_table_hypergate(df_hypergate,dataset = 'large PBMC', cell_type_key = 'cluster cell_type_lvl_1'):
    df_hypergate = df_hypergate.rename(columns={'cell_type': 'celltype'})
    df_hypergate['dataset'] = dataset
    df_hypergate['model'] = 'Hypergate'
    df_hypergate = df_hypergate.pivot_table(
    index=['sample', 'celltype', 'set_size', 'ratio', 'train_ratio', 'time', 'model','dataset'],
    columns='score',
    values='value'
        ).reset_index()
    df_hypergate = df_hypergate.drop(columns=['train_ratio'])
    df_hypergate.columns.name = None
    df_hypergate['cell_type_key'] = cell_type_key
    df_hypergate = df_hypergate.rename(columns={'set_size': 'set'})
    return df_hypergate
```

```python
def plot_scoretype_heatmap(ax, df, score_type, title):
    # Filter by score_type and drop NaNs
    df_sub = df[df['score_type'] == score_type].dropna(subset=['Hypergate'])

    # Determine winner
    df_sub['winner'] = np.where(
        df_sub['ConvexGating'] > df_sub['Hypergate'], 'ConvexGating higher',
        np.where(df_sub['ConvexGating'] < df_sub['Hypergate'], 'Hypergate higher', 'Equal')
    )

    # Count wins per dataset_celltypekey
    count_df = df_sub.groupby(['dataset_celltypekey', 'winner']).size().unstack(fill_value=0)

    # Ensure both model columns exist
    for col in ['ConvexGating higher', 'Hypergate higher']:
        if col not in count_df.columns:
            count_df[col] = 0

    # Compute percentages
    percentage_df = count_df[['ConvexGating higher', 'Hypergate higher']].div(
        count_df.sum(axis=1), axis=0
    ) * 100

    # Create annotation: percentage + count
    annot_df = (
        percentage_df.round(2).astype(str) + '%' +
        " (" + count_df[['ConvexGating higher', 'Hypergate higher']].astype(str) + ")"
    )

    # Plot heatmap
    sns.heatmap(
        percentage_df,
        annot=annot_df.values,
        fmt='',
        cmap='magma_r',
        vmin=0,
        vmax=100,
        ax=ax
    )
    ax.set_title(title)
    ax.tick_params(axis='x', rotation=0)

# Create 3 subplots
fig, axes = plt.subplots(3, 1, figsize=(8, 16))

plot_scoretype_heatmap(axes[0], df_pivot_1to15, 'precision', "Precision - 1:15 ratio")
plot_scoretype_heatmap(axes[1], df_pivot_1to15, 'recall', "Recall - 1:15 ratio")
plot_scoretype_heatmap(axes[2], df_pivot_1to15, 'f1', "F1 - 1:15 ratio")

plt.tight_layout()
plt.savefig(os.path.join(base_save_path_rev11,pre + '_comp_Hypergate_CG_winner_1to15.pdf'),bbox_inches='tight')
plt.show()
```

```python
def _detect_svm_col(df, svm_col):
    if svm_col is not None:
        if svm_col not in df.columns:
            raise KeyError(f"'{svm_col}' not in DataFrame columns.")
        return svm_col
    # Try common variations
    candidates = ["LinearSVM", "Linear SVM", "linear SVM", "SVM_linear", "Linear-SVM"]
    for c in candidates:
        if c in df.columns:
            return c
    raise KeyError("Could not find a Linear SVM column. Provide `svm_col=` explicitly.")

def plot_scoretype_heatmap_linSVM(ax, df, score_type, title, svm_col=None):
    # Figure out which column holds the Linear SVM scores
    svm_col = _detect_svm_col(df, svm_col)

    # Filter by score_type and drop rows missing either model
    df_sub = (
        df[df['score_type'] == score_type]
        .dropna(subset=['ConvexGating', svm_col])
        .copy()
    )

    # Determine winner
    # Labels: ConvexGating higher / Linear SVM higher / Equal
    df_sub['winner'] = np.where(
        df_sub['ConvexGating'] > df_sub[svm_col], 'ConvexGating higher',
        np.where(df_sub['ConvexGating'] < df_sub[svm_col], 'Linear SVM higher', 'Equal')
    )

    # Count wins per dataset_celltypekey
    count_df = df_sub.groupby(['dataset_celltypekey', 'winner']).size().unstack(fill_value=0)

    # Ensure both model columns exist for consistent plotting
    for col in ['ConvexGating higher', 'Linear SVM higher']:
        if col not in count_df.columns:
            count_df[col] = 0

    # Compute percentages for the two models only
    model_cols = ['ConvexGating higher', 'Linear SVM higher']
    percentage_df = count_df[model_cols].div(count_df.sum(axis=1), axis=0) * 100

    # Create annotation: "<pct>% (<count>)"
    annot_df = (
        percentage_df.round(2).astype(str) + '%' +
        " (" + count_df[model_cols].astype(str) + ")"
    )

    # Plot heatmap
    sns.heatmap(
        percentage_df,
        annot=annot_df.values,
        fmt='',
        cmap='magma_r',
        vmin=0,
        vmax=100,
        ax=ax
    )
    ax.set_title(title)
    ax.tick_params(axis='x', rotation=0)

# Example usage (3 subplots)
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# If your SVM column is named differently, pass svm_col="YourColumnName"
plot_scoretype_heatmap_linSVM(axes[0], df_pivot_1to15, 'precision', "Precision - 1:15 ratio", svm_col=None)
plot_scoretype_heatmap_linSVM(axes[1], df_pivot_1to15, 'recall',    "Recall - 1:15 ratio",    svm_col=None)
plot_scoretype_heatmap_linSVM(axes[2], df_pivot_1to15, 'f1',        "F1 - 1:15 ratio",        svm_col=None)
plt.tight_layout()
plt.savefig(os.path.join(base_save_path_rev11, pre + '_comp_LinearSVM_CG_winner_1to15.pdf'), bbox_inches='tight')
plt.show()
```
