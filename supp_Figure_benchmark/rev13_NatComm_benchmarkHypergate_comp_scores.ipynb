```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
import pandas as pd
```

```python
base_path = os.getcwd()
pre = 'rev13_NatComm'
base_save_path_rev10 = 'C:\\Users\\vfriedrich\\projects\\gating\\revision\\NatComm\\rev10_NatComm'
base_save_path_rev12 = 'C:\\Users\\vfriedrich\\projects\\gating\\revision\\NatComm\\rev12_NatComm'
base_save_path_rev13 = 'C:\\Users\\vfriedrich\\projects\\gating\\revision\\NatComm\\rev13_NatComm'
```

```python
df_melted_test = pd.read_csv(os.path.join(base_save_path_rev12,'rev12_NatComm' + '_df_melted_1to1_all.csv'),index_col=0)
```

```python
mean_scores = (
    df_melted_test.groupby(['model', 'score_type', 'dataset_celltypekey'])['score_value']
      .mean()
      .reset_index()
)
```

```python
df_combined_f1 = mean_scores[mean_scores['score_type'] == 'f1'].pivot(
    index='dataset_celltypekey',
    columns='model',
    values='score_value'
).reset_index()
df_combined_f1.columns.name = None
```

```python
df_combined_precision = mean_scores[mean_scores['score_type'] == 'precision'].pivot(
    index='dataset_celltypekey',
    columns='model',
    values='score_value'
).reset_index()
df_combined_precision.columns.name = None
```

```python
df_combined_recall = mean_scores[mean_scores['score_type'] == 'recall'].pivot(
    index='dataset_celltypekey',
    columns='model',
    values='score_value'
).reset_index()
df_combined_recall.columns.name = None
```

```python
df_combined_f1.to_csv(os.path.join(base_save_path_rev13,pre + '_df_combined_mean_f1.csv'))

df_combined_precision.to_csv(os.path.join(base_save_path_rev13,pre + '_df_combined_mean_precision.csv'))

df_combined_recall.to_csv(os.path.join(base_save_path_rev13,pre + '_df_combined_mean_recall.csv'))
```
