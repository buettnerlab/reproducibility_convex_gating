```python
oetjen_all_path = os.path.join('C:\\Users\\vfriedrich\\projects','gating','reproducibility_cleaned','supp_Figure_oetjen','oetjen_all.h5ad')

oetjen_all = sc.read_h5ad(oetjen_all_path)
oetjen_all.obs['cell_type_lvl0'] = oetjen_all.obs['cell_type_lvl0'].astype(str) + "_lvl0"
oetjen_all.obs['cell_type_lvl2'] = oetjen_all.obs['cell_type_lvl2'].astype(str) + "_lvl2"
oetjen_all.obs['cell_type_lvl3'] = oetjen_all.obs['cell_type_lvl3'].astype(str) + "_lvl3"
oetjen_all.obs['cell_type_lvl4'] = oetjen_all.obs['cell_type_lvl4'].astype(str) + "_lvl4"
oetjen_all.obs['cell_type_lvl5'] = oetjen_all.obs['cell_type_lvl5'].astype(str) + "_lvl5"
```

```python
#Without_labels
levels_0 = ['cell_type_lvl0', 'cell_type_lvl2']
levels_1 = ['cell_type_lvl2', 'cell_type_lvl3']
levels_2 = ['cell_type_lvl3', 'cell_type_lvl4']
levels_3 = ['cell_type_lvl4', 'cell_type_lvl5']

annotations_lvl0_lvl2 = oetjen_all.obs[levels_0]
sankey_df_lvl0_lvl2 = annotations_lvl0_lvl2.groupby([levels_0[0], levels_0[1]]).size().reset_index(name='value')
sankey_df_lvl0_lvl2.columns = ['source', 'target', 'value']


annotations_lvl2_lvl3 = oetjen_all.obs[levels_1]
sankey_df_lvl2_lvl3 = annotations_lvl2_lvl3.groupby([levels_1[0], levels_1[1]]).size().reset_index(name='value')
sankey_df_lvl2_lvl3.columns = ['source', 'target', 'value']

annotations_lvl3_lvl4 = oetjen_all.obs[levels_2]
sankey_df_lvl3_lvl4 = annotations_lvl3_lvl4.groupby([levels_2[0], levels_2[1]]).size().reset_index(name='value')
sankey_df_lvl3_lvl4.columns = ['source', 'target', 'value']

annotations_lvl4_lvl5 = oetjen_all.obs[levels_3]
sankey_df_lvl4_lvl5 = annotations_lvl4_lvl5.groupby([levels_3[0], levels_3[1]]).size().reset_index(name='value')
sankey_df_lvl4_lvl5.columns = ['source', 'target', 'value']

combined_sankey_df = pd.concat([annotations_lvl0_lvl2,sankey_df_lvl2_lvl3, sankey_df_lvl3_lvl4, sankey_df_lvl4_lvl5])

labels = list(pd.concat([combined_sankey_df['source'], combined_sankey_df['target']]).unique())
label_indices = {label: i for i, label in enumerate(labels)}

# Convert labels to indices in the dataframe
combined_sankey_df['source'] = combined_sankey_df['source'].map(label_indices)
combined_sankey_df['target'] = combined_sankey_df['target'].map(label_indices)

# Create Sankey plot
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=[''] * len(labels),
    ),
    link=dict(
        source=combined_sankey_df['source'],
        target=combined_sankey_df['target'],
        value=combined_sankey_df['value']
    )
))

#fig.update_layout(title_text="Sankey Diagram of Cell Type Annotations from lvl2 to lvl5", font_size=10)

# Show the figure
fig.show()
```
