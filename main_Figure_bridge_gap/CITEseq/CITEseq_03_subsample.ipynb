```python
file_path = file_path = os.getcwd() 
data_dir = os.path.join("C:\\Users\\vfriedrich\\projects\\gating\\","reproducibility_cleaned",
             "main_Figure_bridge_gap",
             "CITEseq",
            "input_data\\") 
#data_raw_dir = file_path + 'input/' 
sc.settings.figdir = file_path + 'figures/'
today = datetime.date.today().strftime('%y%m%d') #creates a YYMMDD string of today's date
pdata = md.read(f'{data_dir}haniffa21_totalVI_Covid_Healthy_annotated_HLADR+.h5mu/protein')
```

```python
pdata_CD16_CD4_target = pdata[pdata.obs['HLA-DR+ CD16+ T cells'] == 'HLA-DR+ CD16+ CD4+ T cell']
pdata_CD16_CD4_non_target = pdata[pdata.obs['HLA-DR+ CD16+ T cells'] != 'HLA-DR+ CD16+ CD4+ T cell']
sc.pp.subsample(pdata_CD16_CD4_non_target ,n_obs = 15*len(pdata_CD16_CD4_target))
pdata_CD16_CD4_sub_denoised = pdata_CD16_CD4_non_target.concatenate(pdata_CD16_CD4_target)

pdata_CD16_CD8_target = pdata[pdata.obs['HLA-DR+ CD16+ T cells'] == 'HLA-DR+ CD16+ CD8+ T cell']
pdata_CD16_CD8_non_target = pdata[pdata.obs['HLA-DR+ CD16+ T cells'] != 'HLA-DR+ CD16+ CD8+ T cell']
sc.pp.subsample(pdata_CD16_CD8_non_target ,n_obs = 15*len(pdata_CD16_CD8_target))
pdata_CD16_CD8_sub_denoised = pdata_CD16_CD8_non_target.concatenate(pdata_CD16_CD8_target)
```

```python
pdata_CD16_CD4_sub_denoised.write_h5ad(os.path.join(path_output_script,'pdata_CD16_CD4_sub_denoised_HLADR+.h5ad'))
pdata_CD16_CD8_sub_denoised.write_h5ad(os.path.join(path_output_script,'pdata_CD16_CD8_sub_denoised_HLADR+.h5ad'))
```
