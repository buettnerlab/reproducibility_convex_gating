# Create graphics for paper Figure 5

```python
import sys
import os
import scanpy as sc
import pandas as pd
path_helper = ["c:\\","Users","vfriedrich","projects","gating","reproducibility_cleaned","helper"]
sys.path.append(os.path.join(*path_helper))
import helper_reproducibility as hr
import numpy as np
```

```python
pre = 'P06_summary'
start_time = hr.get_time()
```

```python
path_input_data = hr.return_input_data_path() 
path_output_data = hr.return_output_data_path()
path_output_script = os.path.join(path_output_data,pre)
hr.make_path(os.path.join(path_output_script))
```

```python
summary_full = pd.read_csv(os.path.join(os.getcwd(),'output_data','P03_summary','P03_summary_melted_full_strategy.csv'),index_col=0)
summary_2h = pd.read_csv(os.path.join(os.getcwd(),'output_data','P03_summary','P03_summary_melted_2h_first_two_hierachies.csv'),index_col=0)
```

```python
panels = ['full','2h']
levels = ['lvl2','lvl3','lvl4','lvl5']
df_mean = pd.DataFrame(columns=['mean_f1', 'mean_precision', 'mean_recall','level','panel'])
```

```python
panel = '2h'
for panel in panels:
    if panel == 'full':
        for level in levels:
            mean_f1 = np.mean(summary_full[(summary_full['metric'] == 'f1')&(summary_full['level'] == level)]['score']) 
            mean_precision = np.mean(summary_full[(summary_full['metric'] == 'precision')&(summary_full['level'] == level)]['score']) 
            mean_recall = np.mean(summary_full[(summary_full['metric'] == 'recall')&(summary_full['level'] == level)]['score']) 
            df_mean = df_mean.append({'mean_f1': mean_f1,
                            'mean_precision': mean_precision,
                            'mean_recall': mean_recall,
                           'level': level,
                           'panel': panel}, ignore_index=True)
    else:
        for level in levels:
            mean_f1 = np.mean(summary_2h[(summary_2h['metric'] == 'f1')&(summary_2h['level'] == level)]['score']) 
            mean_precision = np.mean(summary_2h[(summary_2h['metric'] == 'precision')&(summary_2h['level'] == level)]['score']) 
            mean_recall = np.mean(summary_2h[(summary_2h['metric'] == 'recall')&(summary_2h['level'] == level)]['score']) 
            df_mean = df_mean.append({'mean_f1': mean_f1,
                            'mean_precision': mean_precision,
                            'mean_recall': mean_recall,
                           'level': level,
                           'panel': panel}, ignore_index=True)
```

```python
#save before plotting
df_mean.to_csv(os.path.join(path_output_script,pre + '_df_mean.csv'))
```
